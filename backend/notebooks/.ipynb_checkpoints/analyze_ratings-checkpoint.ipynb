```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the ratings dataset
ratings = pd.read_csv('../backend/data/raw/ratings.csv')

print(f"Original dataset shape: {ratings.shape}")
```

```python
# Check for missing values
print("\nMissing values:")
print(ratings.isnull().sum())
```

```python
# Basic statistics of ratings
print("\nRatings statistics:")
print(ratings['rating'].describe())
```

```python
# Distribution of ratings
plt.figure(figsize=(10, 6))
sns.histplot(ratings['rating'], bins=10, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Number of ratings per user
user_rating_counts = ratings['userId'].value_counts()
print("\nRatings per user statistics:")
print(user_rating_counts.describe())

plt.figure(figsize=(10, 6))
sns.histplot(user_rating_counts, bins=50, kde=True)
plt.title('Distribution of Ratings per User')
plt.xlabel('Number of Ratings')
plt.ylabel('Number of Users')
plt.show()
```

```python
# Number of ratings per movie
movie_rating_counts = ratings['movieId'].value_counts()
print("\nRatings per movie statistics:")
print(movie_rating_counts.describe())

plt.figure(figsize=(10, 6))
sns.histplot(movie_rating_counts, bins=50, kde=True)
plt.title('Distribution of Ratings per Movie')
plt.xlabel('Number of Ratings')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# Analyze timestamp
ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], unit='s')
ratings['year'] = ratings['timestamp'].dt.year

plt.figure(figsize=(12, 6))
ratings['year'].value_counts().sort_index().plot(kind='bar')
plt.title('Number of Ratings per Year')
plt.xlabel('Year')
plt.ylabel('Number of Ratings')
plt.show()
```

```python
# Suggest filtering criteria
print("\nSuggested filtering criteria:")
print("1. Remove users with very few ratings (e.g., less than 5)")
print("2. Remove movies with very few ratings (e.g., less than 10)")
print("3. Consider removing outlier ratings (e.g., 0.5 and 5.0) if they seem suspicious")
print("4. Consider focusing on a specific time range if there's a significant difference in rating patterns over time")
```

```python
# Apply suggested filters
min_user_ratings = 5
min_movie_ratings = 10

filtered_ratings = ratings[
    ratings['userId'].isin(user_rating_counts[user_rating_counts >= min_user_ratings].index) &
    ratings['movieId'].isin(movie_rating_counts[movie_rating_counts >= min_movie_ratings].index)
]

print(f"\nFiltered dataset shape: {filtered_ratings.shape}")
print(f"Removed {len(ratings) - len(filtered_ratings)} entries")
```

```python
# Save filtered dataset
filtered_ratings.to_csv('../backend/data/processed/filtered_ratings.csv', index=False)
print("\nFiltered dataset saved to '../backend/data/processed/filtered_ratings.csv'")
```
