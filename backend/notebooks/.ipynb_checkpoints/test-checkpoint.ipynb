```python
# Cell 1: Import Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import itertools
from scipy.sparse import csr_matrix
```

```python
# Cell 2: Load data
movies = pd.read_csv('../backend/data/processed/filtered_movies.csv')
ratings = pd.read_csv('../backend/data/processed/filtered_ratings.csv')
```

```python
# Cell 3: Data Analysis
print("Shape of original ratings DataFrame:", ratings.shape)
print("\nFirst few rows of ratings:")
print(ratings.head())

ratings_per_user = ratings.groupby('userId').size()
ratings_per_movie = ratings.groupby('movieId').size()

print("\nRatings per user statistics:")
print(ratings_per_user.describe())

print("\nRatings per movie statistics:")
print(ratings_per_movie.describe())

plt.figure(figsize=(10, 6))
plt.hist(ratings_per_user, bins=50)
plt.title('Distribution of Ratings per User')
plt.xlabel('Number of Ratings')
plt.ylabel('Number of Users')
plt.show()
```

```python
# Cell 4: Create Sparse User-Movie Matrix and Normalize
user_ids = ratings['userId'].unique()
movie_ids = ratings['movieId'].unique()

user_id_map = {id: i for i, id in enumerate(user_ids)}
movie_id_map = {id: i for i, id in enumerate(movie_ids)}

rows = ratings['userId'].map(user_id_map)
cols = ratings['movieId'].map(movie_id_map)
data = ratings['rating']

user_movie_matrix = csr_matrix((data, (rows, cols)), shape=(len(user_ids), len(movie_ids)))

print("Shape of user_movie_matrix:", user_movie_matrix.shape)
print("Sparsity: {:.2%}".format(1 - user_movie_matrix.nnz / (user_movie_matrix.shape[0] * user_movie_matrix.shape[1])))

def normalize_ratings(R):
    R = R.tocsr()  # Ensure the matrix is in CSR format
    user_ratings_mean = np.array(R.sum(axis=1)).flatten() / np.diff(R.indptr)
    R_normalized = R.copy()
    
    for i in range(R.shape[0]):
        start, end = R.indptr[i], R.indptr[i+1]
        R_normalized.data[start:end] -= user_ratings_mean[i]
    
    return R_normalized, user_ratings_mean

user_movie_matrix_normalized, user_ratings_mean = normalize_ratings(user_movie_matrix)

print("\nShape of normalized user_movie_matrix:", user_movie_matrix_normalized.shape)

# Verify normalization
print("\nVerifying normalization:")
for i in range(5):  # Check first 5 users
    user_ratings = user_movie_matrix_normalized[i].data
    if len(user_ratings) > 0:
        print(f"User {i} mean: {user_ratings.mean():.4f}")
    else:
        print(f"User {i} has no ratings")
```

```python
# Cell 5: SVD Model Function
def svd_model(R, k, learning_rate=0.005, reg_param=0.02, n_epochs=100, early_stopping_patience=5):
    n_users, n_items = R.shape
    P = np.random.normal(scale=1./k, size=(n_users, k))
    Q = np.random.normal(scale=1./k, size=(n_items, k))
    
    bu = np.zeros(n_users)
    bi = np.zeros(n_items)
    
    mu = R.data.mean()
    
    best_rmse = float('inf')
    patience_counter = 0
    
    users, items = R.nonzero()
    
    for epoch in range(n_epochs):
        # Shuffle the data
        idx = np.arange(len(users))
        np.random.shuffle(idx)
        
        for i in idx:
            u, i = users[i], items[i]
            r = R[u, i]
            prediction = mu + bu[u] + bi[i] + np.dot(P[u], Q[i].T)
            error = r - prediction
            
            bu[u] += learning_rate * (error - reg_param * bu[u])
            bi[i] += learning_rate * (error - reg_param * bi[i])
            P[u] += learning_rate * (error * Q[i] - reg_param * P[u])
            Q[i] += learning_rate * (error * P[u] - reg_param * Q[i])
        
        # Compute RMSE on a sample
        sample_size = min(len(users), 100000)
        sample_indices = np.random.choice(len(users), sample_size, replace=False)
        sample_users = users[sample_indices]
        sample_items = items[sample_indices]
        
        sample_predictions = mu + bu[sample_users] + bi[sample_items] + np.sum(P[sample_users] * Q[sample_items], axis=1)
        sample_true = R[sample_users, sample_items].A.ravel()
        
        rmse = np.sqrt(np.mean((sample_true - sample_predictions)**2))
        
        print(f"Epoch {epoch}, RMSE: {rmse:.4f}")
        
        if rmse < best_rmse:
            best_rmse = rmse
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= early_stopping_patience:
                print(f"Early stopping at epoch {epoch}")
                break
        
        learning_rate *= 0.9
    
    return P, Q, bu, bi, mu
```

```python
# Cell 6: Prediction Function
def predict(P, Q, bu, bi, mu, user, item):
    return mu + bu[user] + bi[item] + np.dot(P[user], Q[item].T)
```

```python
# Cell 7: Train Final Model
best_params = {
    'k': 10,
    'learning_rate': 0.01,
    'reg_param': 0.01,
    'n_epochs': 100,
    'early_stopping_patience': 5
}

print("Training final model with best parameters:", best_params)
P, Q, bu, bi, mu = svd_model(user_movie_matrix_normalized, **best_params)

print("Final model training complete.")
```

```python
# Cell 8: Calculate and Print RMSE for the Training Data
users, items = user_movie_matrix_normalized.nonzero()
sample_size = min(len(users), 100000)  # Adjust this value based on your memory constraints
sample_indices = np.random.choice(len(users), sample_size, replace=False)
sample_users, sample_items = users[sample_indices], items[sample_indices]

sample_predictions = mu + bu[sample_users] + bi[sample_items] + np.sum(P[sample_users] * Q[sample_items], axis=1)
sample_true = user_movie_matrix_normalized[sample_users, sample_items].A.ravel()

train_rmse = np.sqrt(mean_squared_error(sample_true, sample_predictions))
print(f"Training RMSE: {train_rmse:.4f}")
```

```python
# Cell 9: Recommendation Function
def recommend_movies(user_id, n=10):
    if user_id not in user_id_map:
        print(f"User {user_id} not found in the dataset.")
        return [], []
    
    user_idx = user_id_map[user_id]
    user_ratings = user_movie_matrix[user_idx].toarray().flatten()
    
    unrated_movies = np.where(user_ratings == 0)[0]
    
    if len(unrated_movies) == 0:
        print(f"User {user_id} has rated all available movies.")
        return [], []
    
    raw_predictions = [predict(P, Q, bu, bi, mu, user_idx, movie_idx) for movie_idx in unrated_movies]
    
    # Scale predictions to 1-5 range
    min_pred, max_pred = min(raw_predictions), max(raw_predictions)
    scaled_predictions = [1 + (pred - min_pred) * 4 / (max_pred - min_pred) for pred in raw_predictions]
    
    # Get indices for top and bottom n predictions
    top_n = np.argsort(scaled_predictions)[::-1][:n]
    bottom_n = np.argsort(scaled_predictions)[:n]
    
    inverse_movie_id_map = {i: id for id, i in movie_id_map.items()}
    
    top_recommendations = [(movies.loc[movies['movieId'] == inverse_movie_id_map[unrated_movies[i]]]['title'].iloc[0], scaled_predictions[i]) for i in top_n]
    bottom_recommendations = [(movies.loc[movies['movieId'] == inverse_movie_id_map[unrated_movies[i]]]['title'].iloc[0], scaled_predictions[i]) for i in bottom_n]
    
    return top_recommendations, bottom_recommendations
```

```python
# Cell 10: Updated Example Usage
user_id = 10
top_recommendations, bottom_recommendations = recommend_movies(user_id)

print(f"\nTop 10 recommendations for user {user_id}:")
for movie, score in top_recommendations:
    print(f"{movie}: {score:.2f}")

print(f"\nBottom 5 recommendations for user {user_id}:")
for movie, score in bottom_recommendations[:5]:  # Only display the bottom 5
    print(f"{movie}: {score:.2f}")
```
