```python
# Cell 2: Uppdatera evaluate_model funktionen
def evaluate_model(ratings, user_item_matrix, hybrid_sim, test_size=0.2):
    train, test = train_test_split(ratings, test_size=test_size, random_state=42)
    train_matrix = train.pivot(index='userId', columns='movieId', values='rating').fillna(0)
    
    predictions = []
    actuals = []
    
    for _, row in test.iterrows():
        user_id = row['userId']
        movie_id = row['movieId']
        try:
            pred = predict_rating(user_id, movie_id, user_item_matrix, hybrid_sim)
            if pred is not None:
                predictions.append(pred)
                actuals.append(row['rating'])
        except Exception as e:
            print(f"Error predicting for user {user_id}, movie {movie_id}: {str(e)}")
    
    mae = mean_absolute_error(actuals, predictions)
    rmse = mean_squared_error(actuals, predictions, squared=False)
    
    return mae, rmse
```

```python
# Cell 3: Implementera och utvärdera baseline-modellen
def baseline_model(ratings, test_size=0.2):
    _, test = train_test_split(ratings, test_size=test_size, random_state=42)
    avg_rating = ratings['rating'].mean()
    baseline_rmse = np.sqrt(mean_squared_error(test['rating'], [avg_rating] * len(test)))
    return baseline_rmse

baseline_rmse = baseline_model(ratings)
print(f"Baseline RMSE: {baseline_rmse}")
```

```python
def evaluate_hybrid_recommender(ratings, user_item_matrix, hybrid_sim):
    mae, rmse = evaluate_model(ratings, user_item_matrix, hybrid_sim)
    print(f"Hybrid Recommender - MAE: {mae:.2f}, RMSE: {rmse:.2f}")

    # Jämför med en enkel baseline (genomsnittligt betyg)
    avg_rating = ratings['rating'].mean()
    baseline_mse = mean_squared_error(ratings['rating'], [avg_rating] * len(ratings))
    baseline_rmse = np.sqrt(baseline_mse)
    print(f"Baseline (Average Rating) - RMSE: {baseline_rmse:.2f}")

    # Beräkna förbättring
    improvement = (baseline_rmse - rmse) / baseline_rmse * 100
    print(f"Förbättring jämfört med baseline: {improvement:.2f}%")

    # Skapa y_true och y_pred för att returnera
    test_size = 0.2
    _, test = train_test_split(ratings, test_size=test_size, random_state=42)
    y_true = test['rating'].values
    y_pred = []
    for _, row in test.iterrows():
        pred = predict_rating(row['userId'], row['movieId'], user_item_matrix, hybrid_sim)
        y_pred.append(pred if pred is not None else avg_rating)

    return mae, rmse, baseline_rmse, improvement, y_true, y_pred
```
